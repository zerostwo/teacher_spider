```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
js_urls = ["http://foodsci.jiangnan.edu.cn/szdw/js.htm", "http://foodsci.jiangnan.edu.cn/szdw/js/1.htm", 
        "http://foodsci.jiangnan.edu.cn/szdw/js/2.htm", "http://foodsci.jiangnan.edu.cn/szdw/js/3.htm",
        "http://foodsci.jiangnan.edu.cn/szdw/js/4.htm", "http://foodsci.jiangnan.edu.cn/szdw/js/5.htm"]

fjs_urls = ["http://foodsci.jiangnan.edu.cn/szdw/fjs1.htm", "http://foodsci.jiangnan.edu.cn/szdw/fjs1/6.htm",
       "http://foodsci.jiangnan.edu.cn/szdw/fjs1/5.htm", "http://foodsci.jiangnan.edu.cn/szdw/fjs1/4.htm",
       "http://foodsci.jiangnan.edu.cn/szdw/fjs1/3.htm", "http://foodsci.jiangnan.edu.cn/szdw/fjs1/2.htm",
       "http://foodsci.jiangnan.edu.cn/szdw/fjs1/1.htm"]
```

```python
df = pd.DataFrame(columns=["name", "homepage", "picture", "degree", "research_center", "research_field", "telephone", "mobile_phone", "email", "address"])
for url in js_urls:
    r = requests.get(url)
    r.encoding = r.apparent_encoding
    soup = BeautifulSoup(r.text)
    a = soup.find_all("a", {"style": "color:#454545;font-size:15px;display: inline-block;"})
    df1 = pd.DataFrame(columns=["name", "homepage", "picture", "degree", "research_center", "research_field", "telephone", "mobile_phone", "email", "address"])
    for i in range(len(a)):
        url1 = "http://foodsci.jiangnan.edu.cn/szdw/js/" + a[i].get("href").replace("js/", "")
        info1 = [a[i].get("title"), 
          url1, 
          "http://foodsci.jiangnan.edu.cn" + a[i].img.get("src")]
        # 获取个人信息            
        r1 = requests.get(url1)
        r1.encoding = r1.apparent_encoding
        soup1 = BeautifulSoup(r1.text)
        div2 = soup1.find_all("div", {"class": "col-8"})
        info2 = []
        for j in div2:
            info2.append(j.string)    
        info = info1 + info2
        if len(info2) != 0:
            df1.loc[i] = info
    df = pd.concat([df, df1])

for url in fjs_urls:
    r = requests.get(url)
    r.encoding = r.apparent_encoding
    soup = BeautifulSoup(r.text)
    a = soup.find_all("a", {"style": "color:#454545;font-size:15px;display: inline-block;"})
    df1 = pd.DataFrame(columns=["name", "homepage", "picture", "degree", "research_center", "research_field", "telephone", "mobile_phone", "email", "address"])
    for i in range(len(a)):
        url1 = "http://foodsci.jiangnan.edu.cn/szdw/fjs1/" + a[i].get("href").replace("fjs1/", "")
        info1 = [a[i].get("title"), 
          url1, 
          "http://foodsci.jiangnan.edu.cn" + a[i].img.get("src")]
        # 获取个人信息            
        r1 = requests.get(url1)
        r1.encoding = r1.apparent_encoding
        soup1 = BeautifulSoup(r1.text)
        div2 = soup1.find_all("div", {"class": "col-8"})
        info2 = []
        for j in div2:
            info2.append(j.string)    
        info = info1 + info2
        if len(info2) != 0:
            df1.loc[i] = info
    df = pd.concat([df, df1])
```
